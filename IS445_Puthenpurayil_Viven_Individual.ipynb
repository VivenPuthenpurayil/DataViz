```python
import pandas as pd
import matplotlib.pyplot as plt
```

This is a dataset of all the superbowl winners in every superbowl played up till 2020 so from superwbowl 1 to 54. The dataset includes the date the match was played: string, the superwbowl number: string, the winner: string, the points the winner scored: int, the loser: string, the points the loser scored: int, the MVP: string, the stadium: string, the city: string, and the state: string. We are going to look at this data in reverse so when constructing graphs we can go in the order from the first superbowl to the last one.

Datasource:

Superbowl History 1967 - 2020: https://www.kaggle.com/datasets/timoboz/superbowl-history-1967-2020

License: CC0: Public Domain

```python
df = pd.read_csv('superbowl.csv')
df.head()
```

```python
df_r = df.iloc[::-1]
df_r.head()
```

```python
winner_counts = df_r['Winner'].value_counts()
plt.bar(winner_counts.index[:5], winner_counts.values[:5])
plt.title("Top 5 Superbowl winners")
plt.xlabel("Team")
plt.ylabel("# of wins")
plt.plot()
```

The above graph didn't work because there isn't enough space on the x-axis for all of the names to fit, so to fix this issue I increased the size of the plot for all the names to comfortably fit.

```python
winner_counts = df_r['Winner'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(winner_counts.index[:5], winner_counts.values[:5])
plt.title("Top 5 Superbowl winners")
plt.xlabel("Team")
plt.ylabel("# of wins")
plt.plot()
```

```python
winner_counts = df_r['MVP'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(winner_counts.index[:5], winner_counts.values[:5])
plt.title("Top 5 Superbowl MVP winners")
plt.xlabel("Player")
plt.ylabel("# of MVP wins")
plt.plot()
```

```python
winner_counts = df_r['State'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(winner_counts.index[:5], winner_counts.values[:5])
plt.title("Top 5 Superbowl hosting states")
plt.xlabel("State")
plt.ylabel("# of superbowls hosted")
plt.plot()
```
