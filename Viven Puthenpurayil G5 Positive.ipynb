```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('store_data.csv')
df2 = df[df['Profit'] > 0]
df2 = df2.sort_values(by=['Profit'])
df2 = df2.head(7900)
```

```python
plt.title("Store Profits")
plt.xlabel("Transaction")
plt.ylabel("Profit")
plt.plot(range(len(df2)),sorted(df2['Profit']))
```
